```python
# Process columns with string categorical values

# Sol :: create a LabelEncoder object and fit it to each feature in X

# import preprocessing from sklearn
from sklearn import preprocessing
df_1 = pd.read_csv("DT_data.csv")
df_1.head()
```

```python
# 1. INSTANTIATE
# encode labels with value between 0 and n_classes-1.
le = preprocessing.LabelEncoder()

# 2/3. FIT AND TRANSFORM
# use df.apply() to apply le.fit_transform to all columns
# This is ir-reversible transformation.
# 
df_2 = df_1.apply(le.fit_transform)

df_2.head()
```

```python
df_3 = pd.get_dummies(df_2, prefix_sep="__",
                              columns=df_2.columns[1:])
df_3.head()
```

```python
# One hot encoding
# Steps: create a OneHotEncoder object, and fit it to all of X

# 1. INSTANTIATE
enc = preprocessing.OneHotEncoder()

# 2. FIT
enc.fit(df_2.iloc[:,1:])

# 3. Transform
onehotlabels = enc.transform(df_2.iloc[:,1:]).toarray()
df_4 = pd.DataFrame(onehotlabels.T)
df_4.head()
```
